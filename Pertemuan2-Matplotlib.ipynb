# Matplotlib 2
## Pengenalan pyplot
- pyplot merupakan koleksi atau kumpulan fungsi yang menjadikan Matplotlib dapat bekerja menyerupai MATLAB.
- pyplot API secara umum less-flexible bila dibandingkan object-oriented API.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
plt.plot([2,5,7,11]) # plotting dengan satu buah deret bilangan dalam struktur data list.
plt.ylabel('sumbu y')
plt.show()
```

- Pemanggilan fungsi plot() dapat dilakukan dengan menyertakan hanya sebuah deret bilangan (list/array) ataupun dua buah deret bilangan (list/array).
- Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point untuk nilai pada sumbu y; sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret bilangan terurut dengan nilai dimulai dari nol.

```python
plt.plot([2,3,4,7], [5,4,2,16]) # plotting dengan 2 buah deret bilangan.
plt.show()
```

## Pengaturan format pada plot
- Selain menyertakan dua buah deret bilangan sebagai parametter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting.
- Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB.
- Nilai default untuk parameter ketiga ini adalah 'b-'.
- Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin, ymax]

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') 
"""
Parameter ketiga berfungsi untuk mengatur warna dan jenis ploting, untuk character pertama
pada parameter ketiga berguna untuk menentukan warna dan character kedua berfungsi untuk
menentukan bentuk atau jenis plotting.
Contoh character pertama:
- r = red
- g = green
- b = blue
dst.
Contoh character kedua:
- o = ploting berupa titik.
- - = ploting berupa garis.
"""
plt.axis([0, 6, 0, 20])
# axis berfungsi untuk mengatur rentang plotting yang ingin ditampilkan.
plt.show()
```

## Multi plot dalam satu pemanggilan fungsi plot()

```python
t = np.arange(0.,5.,0.2) 
""" 
Membuat sebuah deret bilangan terurut. 
Parameter pertama untuk bilangan awal.
Parameter kedua untuk batas akhir bilangan.
Parameter ketiga untuk interval atau step antar bilangan yang dibentuk.
"""
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

## Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot()

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

## Plotting dengan keywords

- Matplotlib juga memungkinkan untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs) seperti pada dictionary dan pandas Data Frame.

```python
data = {'a':np.arange(50),
        'c':np.random.randint(0,50,50),
        'd':np.random.randn(50)}
data['b'] = data['a']+10*np.random.randn(50)
data ['d'] = np.abs(data['d']*108)
data
```

```python
plt.scatter('a','b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## Plotting untuk tipe data categorical
- Matplotlib juga mendukung plotting untuk data dengan tipe kategori atau biasa dikenal dengan istilah categorical data atau categorical variables.

```python
names = ['group_a', 'group_b', 'group_c']
values = [1,10,100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.show()
```
